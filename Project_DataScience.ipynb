```python
#Task.1 Reading the data, identifying missing values and inconsistencies 

import pandas as pd
df = pd.read_csv("Group_11_A.csv", delimiter = ",")
display(df)
print("note: Problems with 'Purchase-Date' and 'Proximity'\n")
display(df.select_dtypes(include='object').describe())
display(df.select_dtypes(exclude='object').describe())
print("\nNumber of null values for every variable:-\n")
display(df.isnull().sum())
print("note: Renovate, Contour, Garage, and Swimming have missing values\n")
print("\nIdentifying the types of the data\n")
display(df.dtypes)
print("note: Proximity should be an int not an object")

#Task.1 Handling missing values
# note: Renovate, Contour, Garage, and Swimming have missing values
#Needs fixing

null_columns = df.columns[df.isna().any()]
print("Columns with NaN values are: ", null_columns)



for c in null_columns:
    value = df[c].mode()[0] if df[c].dtype!="float" or df[c].dtype!="int" else df[c].mean()
    df.fillna({c:value}, inplace = True)
    


display(df)
null_columns = df.columns[df.isna().any()]
print("Columns with NaN values are: ", null_columns)
display(df.isnull().sum())

#Task.1 Inconsistency fixing
def fixDate(x):
    date = x.split("-")
    month = date[0]
    year = date[1]
    if(int(year)<100 and int(year)>21):
        year = "19"+str(year)
        x = year+"-"+month
    elif(int(year)<100 and int(year)<21):
            year = "20"+str(year)
            x = year+"-"+month
    xsplit = x.split("-")
    if(xsplit[0].isdigit()):
        return xsplit[1]+"-"+xsplit[0]
    else:
        return x

def separateYear(date):
    date = date.split("-")
    for element in date:
        if(element.isdigit()):
            return int(element)

def separateMonth(date):
    date = date.split("-")
    for element in date:
        if(not element.isdigit()):
            return element
    
        
df["Proximity"]=df["Proximity"].apply(lambda x: int(x.replace("mts",""))) #Remove "mts" such that the type'll be converted to int
df["Purchase-Date"]=df["Purchase-Date"].apply(lambda x : fixDate(x)) #Replace 2 digits years with the proper '19XX' or '20XX' prefix (example: 03-->2003, 78-->1978)


df["Purchase-Year"] = df["Purchase-Date"].apply(lambda x : separateYear(x))
df["Purchase-Month"] = df["Purchase-Date"].apply(lambda x : separateMonth(x))
monthMapping = {"Aug": "August",
                "Sep": "September",
                "Oct": "October",
                "Nov": "November",
                "Dec": "December",
                "Jan": "January",
                "Feb": "February",
                "Mar": "March",
                "Apr": "April",
                "Jun": "June",
                "Jul": "July",
                "May": "May"}
df["Purchase-Month"] = df["Purchase-Month"].apply(lambda x: monthMapping[x] if x in monthMapping.keys() else x)

display(df)
df.info()

#Task.1 Remove outliers
num_columns = df.select_dtypes(exclude="object").columns
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


plt.figure()
sns.boxplot(y=df['Sale-Price'])
plt.show()

for i in range(4):
    Q3, Q1 = np.percentile(df['Sale-Price'], [75 ,25])
    IQR = Q3-Q1
    upperWhisker = Q3+1.5*IQR
    lowerWhisker = Q1-1.5*IQR
    df = df[(df['Sale-Price']>lowerWhisker)&(df['Sale-Price']<upperWhisker)]
    
display(df.info())
plt.figure()
sns.boxplot(y=df['Sale-Price'])
plt.show()

#display(sns.boxplot(y=df['Sale-Price']))
```

```python
import numpy as np
import pandas as pd
import seaborn as sns

df = pd.read_csv('Group_11_B.csv')
display(df)
display(df.describe())

#One hot encode data to use in regression
df = pd.get_dummies(df, columns=['Type','Class','Location','Shape','Access'],drop_first=True) 

#Purchase date to float
#1/12 approximately 0.084 to convert month into accurate representation mimicing data in X1 transaction date regression-4 csv
def date_to_num(row):
    if (row.split('-')[0] == 'January'):
        return  float(row.split('-')[1])  + 1*0.084
    elif (row.split('-')[0] == 'February'):
        return  float(row.split('-')[1]) + 2*0.084
    elif (row.split('-')[0] == 'March'):
        return  float(row.split('-')[1])  + 3*0.084
    elif (row.split('-')[0] == 'April'):
        return  float(row.split('-')[1])  + 4*0.084
    elif (row.split('-')[0] == 'May'):
        return  float(row.split('-')[1])  + 5*0.084
    elif (row.split('-')[0] == 'June'):
        return  float(row.split('-')[1])  + 6*0.084
    elif (row.split('-')[0] == 'July'):
        return  float(row.split('-')[1])  + 7*0.084
    elif (row.split('-')[0] == 'August'):
        return  float(row.split('-')[1])  + 8*0.084
    elif (row.split('-')[0] == 'September'):
        return  float(row.split('-')[1])  + 9*0.084
    elif (row.split('-')[0] == 'October'):
        return  float(row.split('-')[1])  + 10*0.084
    elif (row.split('-')[0] == 'November'):
        return  float(row.split('-')[1])  + 11*0.084
    elif (row.split('-')[0] == 'December'):
        return  float(row.split('-')[1])  + 12*0.084
    else:
        return row

df['Purchase-Date'] = df['Purchase-Date'].apply(date_to_num)

from sklearn.preprocessing import LabelEncoder
#transform yes no na columns to use in regression
encoder = LabelEncoder()
encoder.fit(df.Garage)
df.Garage=encoder.transform(df.Garage)
encoder.fit(df.Swimming)
df.Swimming=encoder.transform(df.Swimming)
encoder.fit(df.Contour)
df.Contour=encoder.transform(df.Contour)

display(df.dtypes)

from sklearn.model_selection import train_test_split
X = df.iloc[:,1:].values 
y = df.iloc[:, 0].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=22)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(np.c_[X_train,y_train])

A_train = scaler.transform(np.c_[X_train,y_train])
X_train = A_train[:,1:]
y_train = A_train[:,0]

A_test = scaler.transform(np.c_[X_test,y_test])
X_test = A_test[:,1:]
y_test = A_test[:,0]

# Regression Analysis: Mean Squared Error Metric
from sklearn.metrics import mean_squared_error

## OLS
from sklearn.linear_model import LinearRegression
reg1 = LinearRegression(fit_intercept=False).fit(X_train, y_train)
y_pred1 = reg1.predict(X_test)
print('The MSE using OLS is:', mean_squared_error(y_test, y_pred1))
with pd.option_context('display.max_rows', 5, 'display.max_columns', None): 
    display(df)
```
